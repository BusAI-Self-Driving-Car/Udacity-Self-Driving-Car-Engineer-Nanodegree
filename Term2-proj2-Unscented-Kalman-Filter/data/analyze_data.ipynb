```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

lidar_data = pd.read_table("lidar_data.txt", header=None)
radar_data = pd.read_table("radar_data.txt", header=None)

## Plot NIS RADAR
output_radar = pd.read_table("output_radar.txt")
#output_radar
timestamp_radar = np.array([output_radar['time_stamp']])
NIS_radar = np.array([output_radar['NIS']])

print ("NIS RADAR")
plt.plot(np.squeeze(timestamp_radar), 7.8 * np.ones((np.squeeze(NIS_radar)).size))
plt.plot(np.squeeze(timestamp_radar), np.squeeze(NIS_radar))
plt.show()

## Plot NIS LIDAR
output_lidar = pd.read_table("output_lidar.txt")
#output_lidar
timestamp_lidar = np.array([output_lidar['time_stamp']])

print ("NIS LIDAR")
NIS_lidar = np.array([output_lidar['NIS']])
plt.plot(np.squeeze(timestamp_lidar), 5.991 * np.ones((np.squeeze(NIS_lidar)).size))
plt.plot(np.squeeze(timestamp_lidar), np.squeeze(NIS_lidar))
plt.show()
```
